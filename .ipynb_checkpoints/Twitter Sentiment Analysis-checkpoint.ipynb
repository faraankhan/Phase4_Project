This is a decent first step, but there are clearly rows that contain the keywords that are mislabled. Use REGEX to search for specific string patterns and relabel them as the correct product_brand.

```python
keyword_patterns = {
    'apple': r'\bapple\b',
    'store': r'\bstore\b',
    'app': r'\bapp\b'
}

def label_product_brand(row):
    labels = []
    for label, pattern in keyword_patterns.items():
        if re.search(pattern, row['tweet']):
            labels.append(label)
    if 'apple' in labels and 'store' in labels:
        return 'Other Apple product or service'
    elif 'apple' in labels and 'app' in labels:
        return 'iPad or iPhone App'
    elif 'apple' in labels:
        return 'Apple'
    else:
        return row['product_brand']
    
df_app['product_brand'] = df_app.apply(label_product_brand, axis=1)
```

# EDA: Tokenization, Lemmatization, Stop Word Removal

```python
stopwords = nltk.corpus.stopwords.words('english')
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

Create a function that cleans tweets to remove unidentifiable characters and (only ASCII characters) and HTML code.

```python
def clean_tweets(text):
    
    text = re.sub(' +', ' ', text) 
    encoded_text = text.encode('ascii', 'ignore') 
    decode_text = encoded_text.decode() 
    clean_text = BeautifulSoup(decode_text, 'lxml').text 
    
    return clean_text
```

Create a similar function that performs the clean_tweets tasks, but also lemmatizes and tokenizes text.

```python
def clean_tweets_model(text):
    
    text = re.sub(' +', ' ', text) 
    encoded_text = text.encode('ascii', 'ignore') 
    decode_text = encoded_text.decode() 
    clean_text = BeautifulSoup(decode_text, 'lxml').text 
    
    text_no_punct = "".join([word for word in clean_text if word not in string.punctuation]) 
    tokens = re.split('\W+', text_no_punct) 
    
    tokens2 = [wnl.lemmatize(word) for word in tokens if word not in stopwords]
    
    return tokens2  
```

```python
df_cleaned['tweet'] = df_cleaned['tweet'].apply(lambda x: clean_tweets(x))
df_cleaned['tokens'] = df_cleaned['tweet'].apply(lambda x: clean_tweets_model(x))
```

Create a couple new columns in the cleaned dataframe that count the tweet length and the number of mentions in a tweet. These features are inherent to twitter (max character length and the social component of tweeting @ users), so their values may reflect whether a tweet contains a positive or negative sentiment. 

```python
def length_of_tweet(text):
    no_punc = "".join([x for x in text if x not in string.punctuation])
    return len(no_punc.split())

def mentions(text):
    counter = 0
    for x in text:
        if '@' in x:
            counter += 1
        else:
            continue
    return counter

```

```python
df_cleaned['length'] = df_cleaned['tweet'].apply(length_of_tweet)
df_cleaned['mentions'] = df_cleaned['tweet'].apply(mentions)
```

# Building Models

We will begin by making the data binary - we will combine negative and neutral sentiments and classify them as 0, and will classify positive sentiment as 1. This will help with model performance as well as address the dramatic class imbalance issue between positive and negative tweets. From a business perspective, a company is likely to take action if sentiment about its products or brands are negative or even neutral (in an attempt to push neutral minded consumers to the positive bucket to create brand loyalty and new business).  

```python
label_mapping = {
    'Positive': 1,
    'Negative': 0,
    'Neutral': 0}

df_cleaned['sentiment_label'] = df_cleaned['sentiment'].map(label_mapping)
```

```python
df_cleaned.head()
```

```python
df_cleaned['sentiment_label'].value_counts()
```

Note that there is still a class imbalance of roughly 60/40, so we may need to over or under sample accordingly. Begin by TFIDing the tokens and creating training and testing data.

```python
tfidf_vec_bin = TfidfVectorizer()
X_train, X_test, y_train, y_test = train_test_split(df_cleaned.drop(labels=['sentiment_label'], axis=1), 
                                                    df_cleaned['sentiment_label'], 
                                                    test_size=0.3, stratify=df_cleaned['sentiment_label'],
                                                    random_state=3)
```

```python
X_train['filtered_tokens_str'] = X_train['filtered_tokens'].apply(lambda tokens: ' '.join(tokens))
X_test['filtered_tokens_str'] = X_test['filtered_tokens'].apply(lambda tokens: ' '.join(tokens))

# Create and fit the TfidfVectorizer
tfidf_vec_bin = TfidfVectorizer(binary=True)
X_train_sparse = tfidf_vec_bin.fit_transform(X_train['filtered_tokens_str'])
X_test_sparse = tfidf_vec_bin.transform(X_test['filtered_tokens_str'])
```

```python
X_train['length'] = X_train['length'].map(lambda x: np.log(x))
X_test['length'] = X_test['length'].map(lambda x: np.log(x))

scaler = MinMaxScaler()
train_af_scaled = pd.DataFrame(scaler.fit_transform(X_train[['mentions', 'length']]))
train_af_scaled.columns = ['mentions_scaled', 'length_log_scaled']

test_af_scaled = pd.DataFrame(scaler.transform(X_test[['mentions', 'length']]))
test_af_scaled.columns = ['mentions_scaled', 'length_log_scaled']
```

```python
X_train_df = pd.DataFrame(X_train_sparse.toarray())
X_train_df.columns = tfidf_vec_bin.get_feature_names_out()
X_train_af = pd.concat([X_train_df, train_af_scaled], axis=1)


X_test_df = pd.DataFrame(X_test_sparse.toarray())
X_test_df.columns = tfidf_vec_bin.get_feature_names_out()
X_test_af = pd.concat([X_test_df, test_af_scaled], axis=1)
```

```python
def model_performance(model, xtrain, xtest, ytrain, ytest, cross_val_var):
    print(f'Training Accuracy: {model.score(xtrain, ytrain)}')
    print(f'Cross Validation Accuracy: {np.mean(cross_val_var)}')
    print(f'Testing Accuracy: {model.score(xtest, ytest)}')
    print(f'Precision: {precision_score(ytest, model.predict(xtest), average="weighted")}')
    print(f'Recall: {recall_score(ytest, model.predict(xtest), average="weighted")}')
    print(f'F1 Score: {f1_score(ytest, model.predict(xtest), average="weighted")}')
```

## Multinomial Bayes

```python
mnb = MultinomialNB()
mnb.fit(X_train_sparse, y_train)
```

```python
mnb_cv_mean = np.mean(cross_val_score(mnb, X_train_sparse, y_train, cv=5, scoring='accuracy', n_jobs=-1))
```

```python
model_performance(mnb, X_train_sparse, X_test_sparse, y_train, y_test, mnb_cv_mean)
```

```python
print(classification_report(y_test, mnb.predict(X_test_sparse)))
```

```python
y_pred = mnb.predict(X_test_sparse)

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(6, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Greens)
plt.title('Confusion Matrix')
plt.colorbar()

classes = np.unique(y_test)
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, cm[i, j], horizontalalignment='center', color='white' if cm[i, j] > cm.max() / 2 else 'black')

plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
